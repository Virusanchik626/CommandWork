```python
import nbformat
from pathlib import Path

# Путь к папке с ноутбуками
WORK_DIR = Path(r"C:\Users\almaz\Downloads\works")

# Порядок файлов для объединения
notebooks_order = [
    "1_Задание_Итог.ipynb",
    "2_3_Задание_Итог.ipynb",
    "4_задание.ipynb"
]

# Выходной файл
output_file = WORK_DIR / "all_report.ipynb"

# Создаем новый пустой ноутбук
merged_nb = nbformat.v4.new_notebook()

# Добавляем заголовок
merged_nb.cells.append(nbformat.v4.new_markdown_cell(
    "# 📊 ИТОГОВЫЙ ОТЧЁТ\n\n"
    "## Объединённый ноутбук из всех заданий\n\n"
    "---\n"
))

# Объединяем ноутбуки по порядку
for notebook_name in notebooks_order:
    notebook_path = WORK_DIR / notebook_name
    
    if notebook_path.exists():
        print(f"📎 Добавляю: {notebook_name}")
        
        # Читаем текущий ноутбук
        nb = nbformat.read(notebook_path, as_version=4)
        
        # Добавляем разделитель
        merged_nb.cells.append(nbformat.v4.new_markdown_cell(
            f"\n---\n## 📁 Файл: `{notebook_name}`\n---\n"
        ))
        
        # Добавляем все ячейки
        merged_nb.cells.extend(nb.cells)
        
        # Добавляем разделитель после
        merged_nb.cells.append(nbformat.v4.new_markdown_cell("\n---\n"))
    else:
        print(f"⚠️  Файл не найден: {notebook_name}")

# Сохраняем результат
nbformat.write(merged_nb, output_file)

print(f"\n✅ Готово!")
print(f"📄 Итоговый файл: {output_file}")
print(f"📊 Всего ячеек: {len(merged_nb.cells)}")
```
